```python
%matplotlib inline
import requests
import pandas as pd
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt

def clean_closed_blocks_cell(s):
    m,M = s.split(' - ')
    return (int(m) + int(M)) / 2
    
def clean_df(df):
    df['closed blocks'] = [clean_closed_blocks_cell(x) for x in df['closed blocks']]
    return df

def plot_2d_heatmap(df, r_axis_key, c_axis_key, val_axis_key):
    mat_df = df.groupby([c_axis_key, r_axis_key])[[val_axis_key]].mean()
    y_ticks =  [list(mat_df.index.levels[1])[0]] + list(mat_df.index.levels[1])
    x_ticks = [list(mat_df.index.levels[0])[0]] + list(mat_df.index.levels[0])
    mat_df = mat_df.unstack(level=c_axis_key)
    fig = plt.figure(figsize=(5, 4))
    ax = fig.add_subplot(1,1,1)
    ax.set_title("avg. "+ val_axis_key)
    ax.set_ylabel(r_axis_key)
    ax.set_xlabel(c_axis_key)
    ax.set_xticklabels(x_ticks)
    ax.set_yticklabels(y_ticks)
    ax.matshow(mat_df.values, vmin=0, vmax=100, cmap='jet')
    for i in range(mat_df.values.shape[0]):
        for j in range(mat_df.values.shape[1]):
            ax.annotate('%.3f' % (mat_df.values[i,j]),xy=(j-0.2,i)) 
    return mat_df

class HTMLTableParser:
    """
    Adapted from http://srome.github.io/Parsing-HTML-Tables-in-Python-with-BeautifulSoup-and-pandas/
    """
   
    def parse_url(self, url, local=True):
        if local:    
            with open(url, 'rb') as f:
                table_text = f.read()
        else:
            response = requests.get(url)
            table_text = response.text
            
        soup = BeautifulSoup(table_text, 'lxml')
        return [(i,self.parse_html_table(table))\
                for i,table in enumerate(soup.find_all('table'))]  

    def parse_html_table(self, table):
        n_columns = 0
        n_rows=0
        column_names = []

        # Find number of rows and columns
        # we also find the column titles if we can
        for row in table.find_all('tr'):
            
            # Determine the number of rows in the table
            td_tags = row.find_all('td')
            if len(td_tags) > 0:
                n_rows+=1
                if n_columns == 0:
                    # Set the number of columns for our table
                    n_columns = len(td_tags)
                    
            # Handle column names if we find them
            th_tags = row.find_all('th') 
            if len(th_tags) > 0 and len(column_names) == 0:
                for th in th_tags:
                    column_names.append(th.get_text())

        # Safeguard on Column Titles
        if len(column_names) > 0 and len(column_names) != n_columns:
            raise Exception("Column titles do not match the number of columns")

        columns = column_names if len(column_names) > 0 else range(0,n_columns)
        df = pd.DataFrame(columns = columns,
                          index= range(0,n_rows))
        row_marker = 0
        for row in table.find_all('tr'):
            column_marker = 0
            columns = row.find_all('td')
            for column in columns:
                df.iat[row_marker,column_marker] = column.get_text()
                column_marker += 1
            if len(columns) > 0:
                row_marker += 1
                
        # Convert to float if possible
        for col in df:
            try:
                df[col] = df[col].astype(float)
            except ValueError:
                pass
        
        return df
```

```python
url = "oa-benchmark-output/benchmark_1.htm"
html_parser = HTMLTableParser()
table_to_display = 1 # out of all tables on benchmark page (0,1,..)
tables = html_parser.parse_url(url, local=True)
results_df = clean_df(tables[table_to_display][1])
    
    
```

```python
mat_df = plot_2d_heatmap(results_df, '# Nodes', 'threshold size', 'closed blocks')
```
